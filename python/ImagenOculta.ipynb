```python
#se importan las librerias
import cv2
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
#Se lee la informaci√≥n de los archivos en un np.array de numero complejos
Ptxt= np.loadtxt("a.txt",delimiter=',',dtype='complex')
Pcsv= np.loadtxt("a.csv",delimiter=',',dtype='complex')
Pdat= np.loadtxt("a.dat",delimiter=',',dtype='complex')
```

```python
#Se grafica la parte Real, imaginaria y el angulo en la informaci√≥n del archivo a.txt
fig, ax = plt.subplots(1,4,figsize=(15, 15))
ax[0].set_title("np.imag(Ptxt)")
ax[0].imshow(np.imag(Ptxt))
ax[1].set_title("np.real(Ptxt)")
ax[1].imshow(np.real(Ptxt))
ax[2].set_title("np.angle(Ptxt)")
ax[2].imshow(np.angle(Ptxt))
ax[3].set_title("Superposici√≥n")
ax[3].imshow(np.angle(Ptxt)+np.imag(Ptxt)+np.real(Ptxt))
fig.tight_layout()
plt.show()
```

```python
#Se grafica la parte Real, imaginaria y el angulo en la informaci√≥n del archivo a.csv
fig, ax = plt.subplots(1,4,figsize=(15, 15))
ax[0].set_title("np.imag(Pcsv)")
ax[0].imshow(np.imag(Pcsv))
ax[1].set_title("np.real(Pcsv)")
ax[1].imshow(np.real(Pcsv))
ax[2].set_title("np.angle(Pcsv)")
ax[2].imshow(np.angle(Pcsv))
ax[3].set_title("Superposici√≥n")
ax[3].imshow(np.angle(Pcsv)+np.imag(Pcsv)+np.real(Pcsv))
fig.tight_layout()
plt.show()
```
