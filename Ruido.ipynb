```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
ruido=cv2.imread("b.png",cv2.IMREAD_UNCHANGED)
furier=cv2.imread("Furier.png",cv2.IMREAD_UNCHANGED)
real=cv2.imread("real.png",cv2.IMREAD_UNCHANGED)
imag=cv2.imread("imag.png",cv2.IMREAD_UNCHANGED)
```

```python
plt.imshow(ruido, cmap="gray")
plt.show()
np.shape(ruido)
```

```python
tf=(np.fft.fft2(ruido))
tf1=np.fft.fftshift(tf)
eso=np.abs(tf1)
plt.imshow(eso,)
plt.show()
plt.imshow(np.log(eso),cmap="gray")
plt.show()
```
