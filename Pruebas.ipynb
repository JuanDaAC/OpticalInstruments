```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
mago=cv2.cvtColor((cv2.imread('Captura.jpg')), cv2.COLOR_BGR2RGB)
mago6=cv2.imread("Captura.jpg",cv2.IMREAD_GRAYSCALE)

mago7=cv2.imread('c.jpg',cv2.IMREAD_GRAYSCALE)
mago1=cv2.imread("c.jpg",cv2.IMREAD_UNCHANGED)

mago2=cv2.imread("b.png",cv2.IMREAD_UNCHANGED)

mago3=cv2.imread("a.csv",cv2.IMREAD_UNCHANGED)

mago4=cv2.imread("a.txt",cv2.IMREAD_UNCHANGED)

mago5=cv2.imread("a.dat",cv2.IMREAD_UNCHANGED)
```

```python
plt.imshow(mago7)
plt.show()
magor=cv2.copyMakeBorder(mago6,312,581,709,949,cv2.BORDER_CONSTANT)
plt.imshow(magor)
plt.show()
np.shape(magor)
```

```python
evil=np.fft.fftshift(np.fft.fftn(magor))
evil2=np.fft.fftshift(np.fft.fftn(mago7))
plt.imshow(np.log(np.abs(evil).conjugate()))
plt.show()
plt.imshow(np.log(np.abs(evil2)))
plt.show()
```

```python
nuevo=(evil.conjugate())*evil2
solucion1=np.fft.ifftshift(np.fft.ifftn(nuevo))
nsolucion1=np.fft.ifftn(np.fft.ifftshift(nuevo))
plt.imshow(np.log(np.abs(nuevo)))
plt.show()
plt.imshow(np.abs(nuevo))
plt.show()
plt.imshow(((np.abs(solucion1))))
plt.show()
plt.imshow((np.abs(nsolucion1)))
plt.show()

cv2.imshow("prueba",((np.abs(nsolucion1))))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
