```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#leemos la imagen 
img=cv2.imread('punto3.png',cv2.IMREAD_GRAYSCALE)
```

```python
#Calculamos la transformada de fourier de la imagen

FFT_img=np.fft.fftn(img)

#desplazamos al origen el cero de frecuencias la transformada de fourier
FFT_img_shift=np.fft.fftshift(FFT_img)

#calculamos la magnitud de la transformada de fourier
magnitude_response=np.log(np.abs(FFT_img_shift)**2)
```

```python
#Imagen de entrada
plt.imshow(img)
```

```python
#transformada de fourier imagen de salida
plt.imshow(magnitude_response)
```
