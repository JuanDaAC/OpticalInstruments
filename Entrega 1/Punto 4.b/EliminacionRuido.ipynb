```python
#Se importan las librerias a implementar
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Se lee la imagen b.png
ruido=cv2.imread("b.png",cv2.IMREAD_UNCHANGED)
plt.imshow(ruido,cmap="gray")
plt.show()
```

```python
#Se saca la tranformada de forier y se grafica el logaritmo de su intensidad
tf=(np.fft.fft2(ruido))
tf1=np.fft.fftshift(tf)
intensidad=np.abs(tf1)
plt.imshow(np.log(intensidad))
plt.show()
```

```python
#Se crea una mascara con fodo de 1 y dos circulos de 0 en la posición de la frecuencias correspondientes al ruido periodico
mask=np.ones((768,768),dtype="uint8")
cv2.circle(mask,(415,371),12,0,-1)
cv2.circle(mask,(353,397),10,0,-1)
plt.title("Mascara")
plt.imshow(mask,cmap="gray")
plt.show()
#Se multiplica la mascara por la tranformada de fourier de la imagen original 
prueba= mask*tf1
plt.title("Log(Intensidad)*Mascara")
plt.imshow(np.log(np.abs(prueba+1))[300:470, 300:470])
plt.show()
plt.title("Resultado")
#Se Grafica el resultado de la mascara por la tranformada de fourier de la imagen original 
plt.imshow((np.abs((np.fft.ifft2(prueba)))),cmap="gray")
plt.show()
```
